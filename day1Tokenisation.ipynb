```python
import nltk
#nltk.download('punkt')
data="hello bithces vaibhav here. hey sal i'm really missing you yar. learning on gr8 learning "

tokens=nltk.sent_tokenize(data)
print(tokens)
```

```python
tokens=nltk.word_tokenize(data)
print(tokens)
```
